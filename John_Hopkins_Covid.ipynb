```python
import pandas as pd
```

```python
corona_confirmed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
corona_confirmed_df.head()
```

```python
corona_deaths_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
corona_deaths_df.head()
```

```python
corona_recovered_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
corona_recovered_df.head()
```

```python
corona_confirmed_df = corona_confirmed_df.melt(
    id_vars=['Province/State', 'Country/Region','Lat',
            'Long'])
```

```python
corona_confirmed_df.head()
```

```python
corona_deaths_df = corona_deaths_df.melt(
    id_vars=['Province/State', 'Country/Region','Lat',
            'Long'])
```

```python
corona_deaths_df.head()
```

```python
corona_recovered_df = corona_recovered_df.melt(
    id_vars=['Province/State', 'Country/Region','Lat',
            'Long'])
```

```python
corona_recovered_df.head()
```

```python
corona_confirmed_df = corona_confirmed_df.rename({
    'variable':'Date',
    'value':'Confirmed'}, axis='columns')
```

```python
corona_confirmed_df.head()
```

```python
corona_deaths_df = corona_deaths_df.rename({
    'variable':'Date',
    'value':'Deaths'}, axis='columns')
```

```python
corona_deaths_df.head()
```

```python
corona_recovered_df = corona_recovered_df.rename({
    'variable':'Date',
    'value':'Recovered'}, axis='columns')
```

```python
corona_recovered_df.head()
```

```python
combined_df = [corona_confirmed_df, corona_deaths_df,corona_recovered_df]

combined_df = [df.set_index(['Province/State',
                            'Country/Region',
                            'Lat','Long', 'Date']) for 
              df in combined_df]
```

```python
combined_df = combined_df[0].join(combined_df[1:])
```

```python
combined_df.head()
```

```python
combined_df = combined_df.reset_index()
```

```python
combined_df.head()
```

```python
combined_df.info()
```

```python
combined_df[['Lat', 'Long', 'Confirmed', 'Deaths', 'Recovered']] = combined_df[['Lat', 'Long', 'Confirmed', 'Deaths', 'Recovered']].apply(pd.to_numeric)
```

```python
combined_df.dtypes
```

```python
combined_df.Date = combined_df.Date.apply(pd.to_datetime)
```

```python
combined_df.dtypes
```

```python
combined_df.head()
```

```python
combined_df = combined_df.rename({
    'Province/State': 'State',
    'Country/Region': 'Country'
},axis=1)
```

```python
combined_df.head()
```

```python
combined_df.shape
```

```python
combined_df.query("Country == 'India'")
```

```python
combined_df.query("Country == 'China'")
```

```python
combined_df['Active'] = combined_df['Confirmed'] - combined_df['Recovered']
```

```python
max_df = combined_df[combined_df.Date==combined_df['Date'].max()]
```
